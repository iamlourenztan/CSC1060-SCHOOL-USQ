## LINEAR REGRESSION

```python
df.head()
```

```python
df=df[['Area', 'Sensing Range', 'Transmission Range', 'Number of Sensor nodes', 'Number of Barriers']]
```

```python
df
```

```python
X = df['Number of Sensor nodes']
y = df['Number of Barriers']
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.xlabel('Number of Sensor nodes')
plt.ylabel('Transmission Range')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state = 23)
```

```python
X_train
```

```python
import numpy as np
X_train = np.array(X_train).reshape(-1, 1)
X_train
```

```python
X_test = np.array(X_train).reshape(-1, 1)
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
c = lr.intercept_
c
```

```python
m = lr.coef_
m
```

```python
Y_pred_train = m*X_train + c
Y_pred_train.flatten()
```

```python
y_pred_train1= lr.predict(X_train)
y_pred_train1
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train)
plt.plot(X_train, y_pred_train1, color='red')
plt.xlabel('Number of Sensor nodes')
plt.ylabel('Transmission Range')
```

```python
y_pred_test1= lr.predict(X_test)
y_pred_test1
```

```python
print(len(X_train), len(y_train))
print(len(X_test), len(y_test))
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred_test1, color='red')
plt.xlabel('Number of Sensor nodes')
plt.ylabel('Transmission Range')
```

## RANDOM FOREST REGRESSION

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df
```

```python
df.head
```

```python
df.dropna(inplace = True)
```

```python
df = df.drop(['Area', 'Sensing Range', 'Transmission Range'], axis = 1)
xVars = df.drop('Number of Barriers', axis = 1)
yVars = df[['Number of Barriers']]
xTrain, xValid, yTrain, yValid = train_test_split(xVars, yVars, test_size = 0.3, random_state = 42)
```

```python
print(xTrain.shape)
print(df.shape)
```

```python
regressor = RandomForestRegressor(n_estimators = 100, random_state = 42)
```

```python
regressor.fit(xTrain, yTrain)
```

```python
y_pred = regressor.predict(xValid)
```

```python
y_pred
```

```python
y_pred = pd.DataFrame(y_pred, columns =['yPredict'])
```

```python
y_pred
```

```python
yValid
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(yValid, y_pred, color = 'red', label = 'Comparison of Prediction between Actual & Prediction Data')
plt.legend()
plt.grid()
plt.title('Random Forest Regression')
plt.xlabel('Prediction Data')
plt.ylabel('Actual Data')
plt.show
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(yValid, y_pred)
print(f'R-squared (R2) score: {r2}')
```
